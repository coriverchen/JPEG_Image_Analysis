```python
import numpy as np
N = 8;
c = np.zeros((8,8))  
alpha = 1.0/np.sqrt(N)
for k in range(N):
    for n in range(N):
        if k == 0:
            c[k,n] = alpha;
        else:
            c[k,n] = alpha * np.cos(np.pi * (2 * n + 1) * k / (2*N));
    alpha = np.sqrt(2/N);
```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(N, N,dpi=600,figsize=(8,8),subplot_kw={'xticks': [], 'yticks': []})

with plt.style.context(['science', 'ieee']):
    for m in range(N):
        plt.subplots_adjust(wspace = .1)
        for n in range(N):
            B = c[[m],:].T.dot(c[[n],:])
            ax[m,n].imshow(B,cmap ='gray',vmin=Ball.min(),vmax=Ball.max())
            plt.subplots_adjust(hspace = .1)
            
plt.savefig("DCTbasis.pdf",bbox_inches='tight')
plt.show()
```
